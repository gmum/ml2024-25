### Zadania domowe
1. Na początku semestru należy założyć prywatne repozytorium na GitHubie przeznaczone na rozwiązane notebooki oraz dać dostęp do repozytorium prowadzącemu ćwiczenia.
2. W trakcie zajęć studenci otrzymają do wypełnienia notebook z ćwiczeniami. Notebook z danych zajęć należy uzupełnić do piątku 23:59 dwa tydzień po otrzymaniu zadania oraz wysłać prowadzącemu mejla z listą wykonanych zadań oraz link do prywatnego repozytorium z rozwiązaniami. Mejl powinien mieć następującą formę:
    * Tytuł: ML_<nr notebooka>_<imię>_<nazwisko>, np. ML_01_Jan_Kowalski,
    * Treść: lista wykonanych zadań (np. 1, 3, 5), link do prywatnego repozytorium z rozwiązanym notebookiem oraz załączoną kopię tego notebooka, która będzie służyła prowadzącemu do weryfikacji rozwiązań.
3. Przyznawanie punktów:
    * Punkty za rozwiązany notebook będą przyznawane w przeciągu tygodnia od deadline'u na wysłanie rozwiązań. Punkty przyznawane będą na podstawie listy zadeklarowanych zadań. 
    * Prowadzący będzie wyrywkowo weryfikował, czy zadeklarowane zadania zostały rzeczywiście poprawnie rozwiązane. 
    * W przypadku, gdy student zadeklarował zadanie, którego nie rozwiązał, rozwiązał je z rażącymi błędami, lub rozwiązał niesamodzielnie, otrzyma 10 punktów karnych, które zostaną odjęte do oceny końcowej z ćwiczeń (skala 0-100). 
    * Za drugą w semestrze niezgodną z prawdą deklarację otrzyma 20 puntków karnych, a za trzecią - 50 punktów karnych (co oznacza niezdanie przedmiotu). 
    * W przypadku, gdy w zadeklarowanym rozwiązaniu pojawią się drobne błędy lub niedociągniecią, punkty za zadanie zostaną odpowiednio pomniejszone, bez przyznawania punktów karnych.
    * W przypadku wątpliwości bądź problemów z rozwiązaniem zadania, proszę o konsultację z prowadzącym w terminie podanym w README.md.
4. Rozwiązania zadań muszą spełniać następujące kryteria:
    - kod musi być czytelny i zwięzły (w razie wątpliwości proszę stosować się do standardu PEP 8),
    - kod musi być przetestowany i poprawnie działać,
    - jeśli do zadania dodane są asserty/testy, kod musi je przechodzić,
    - rozwiązanie musi zostać wysłane w terminie,
    - *komórki oddanego notebooka muszą być wykonane z widocznym rezultatem (prowadzący nie będzie uruchamiał notebooków).*
5. Na zajęciach student może być poproszony o przedstawienie fragmentu lub całości zadeklarowanego rozwiązania i wytłumaczenie działania kodu. W przypadku, gdy student nie będzie w stanie wytłumaczyć działania kodu, będzie traktowane to jako przesłanka do tego, że zadanie zostało rozwiązane niesamodzielnie.
6. Punkty z każdego notebooka (poza "01_zasady_wstep", za któregeo nie otrzymuje się żadnych punktów) będą skalowane do przedziału 0-10 punktów.
7. Suma punktów ze wszystkich notebooków będzie przeskalowana do 80 punktów.
8. Za krótkie presentacji w trakcie ćwiczeń można otrzymać 10 punktów. W wyjątkowych przypadkach prowadzący może ocenić prezentację powyżej 10 punktów.
9. Tematy prezentacji muszą zostać zaakceptowany przez prowadzącego.
10. Tematy prezentacji należy zgłaszać z dwutygodniowym wyprzedzeniem

Osoby dołączające do zajęć w późniejszym terminie muszą liczyć się z faktem, iż nie mogą otrzymać punktów za zaległe zadania.

## Ocena końcowa

Studenci, którzy otrzymali pozytywną ocenę z ćwiczeń (co najmniej 3.0) mają prawo i obowiązek przystąpić do egzaminu pisemnego. Egzamin końcowy w pierwszym terminie przeprowadzony będzie w formie pisemnej. Ocena końcowa z przedmiotu jest oceną łączną z ćwiczeń i egzaminu. Za egzamin można otrzymać 0-100 pkt.

Punktacja końcowa = 0.4 * punkty z egzaminu + 0.6 * punkty z ćwiczeń

* [76-86) pkt: 4
* [86-94) pkt: 4,5
* [94-100) pkt: 5
